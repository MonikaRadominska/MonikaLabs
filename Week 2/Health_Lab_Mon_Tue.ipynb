```python
#Steps to take to clean up the data:
#Rename the column names, standardise data sets, abbreviate the names,
#and sort to match all data sets
```

```python
def rename_columns(x):
    x.rename(columns={'Customer':'customer','ST':'state', "State":"state", 
                          'GENDER':'gender', "Gender":"gender","Education":"education", 
                          "Customer Lifetime Value":"clv","Income": "income",
                          "Monthly Premium Auto":"monthly_premium", "Number of Open Complaints":"complaints",
                           "Policy Type":"policy_type", "Vehicle Class":"vehicle_class",
                            "Total Claim Amount":"total_claim"}, inplace=True )
    return x
```

```python
rename_columns(df1) #running the function above
```

```python
rename_columns(df2)
```

```python
rename_columns(df3)
```

```python
# sort
def sort_columns(x):
    x = x.reindex(columns=['customer', 'state', 'gender', 'education', 'income', 'vehicle_class', 'monthly_premium', 'policy_type', 'complaints', 'total_claim', 'clv'])
    return x
```

```python
df1 =sort_columns(df1)
```

```python
df2 = sort_columns(df2)
```

```python
df3 = sort_columns(df3)
```

```python
#concatinating the data sets
combined = [df1, df2, df3]
df = pd.concat(combined)
df
```

```python
#function to drop columns, dropping customer columns

def drop_columns(x):
    x.drop(columns=["customer"], inplace=True)
    return x
```

```python
df = df.drop_duplicates() #removing duplicates
df
```

```python
df = drop_columns(df)
df
```

Standardizing the states column: Bucketing the data - Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central

```python
#checking the data in the state column to know which names/state abbreviations to standardize
df.state.value_counts()
```

```python
#decided to use a function to change the states
def clean_states(x):
    if x in ['Cali', 'California']:
        return 'West Region'
    elif x in ['AZ', 'Arizona','Nevada']:
        return 'Central'
    elif x in ['WA', 'Washington']:
        return 'East'
    elif x in ['Oregon']:
        return 'North West'
    else:
        return x
```

```python
#using the function to clean the state names and standardize and using variable override to store the changes in the dataframe var
df['state']= df['state'].apply(clean_states)
```
