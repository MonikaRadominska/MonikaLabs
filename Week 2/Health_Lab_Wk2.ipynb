```python
df1 = pd.read_csv('file1.csv')
```

```python
#Steps to take to clean up the data:
#Rename the column names, standardise data sets, abbreviate the names,
#and sort to match all data sets
```

```python
def rename_columns(x):
    x.rename(columns={'Customer':'customer','ST':'state', "State":"state", 
                          'GENDER':'gender', "Gender":"gender","Education":"education", 
                          "Customer Lifetime Value":"clv","Income": "income",
                          "Monthly Premium Auto":"monthly_premium", "Number of Open Complaints":"complaints",
                           "Policy Type":"policy_type", "Vehicle Class":"vehicle_class",
                            "Total Claim Amount":"total_claim"}, inplace=True )
    return x
```

```python
rename_columns(df1) #running the function above
```

```python
rename_columns(df2)
```

```python
rename_columns(df3)
```

```python
# sort
def sort_columns(x):
    x = x.reindex(columns=['customer', 'state', 'gender', 'education', 'income', 'vehicle_class', 'monthly_premium', 'policy_type', 'complaints', 'total_claim', 'clv'])
    return x
```

```python
df1 =sort_columns(df1)
```

```python
df2 = sort_columns(df2)
```

```python
df3 = sort_columns(df3)
```

```python
#concatinating the data sets
combined = [df1, df2, df3]
df = pd.concat(combined)
df
```

```python
#function to drop columns, dropping customer columns

def drop_columns(x):
    x.drop(columns=["customer"], inplace=True)
    return x
```

```python
df = df.drop_duplicates() #removing duplicates
df
```

```python
df = drop_columns(df)
df
```

```python
#check which values are missing in the gender column
len(df[df['gender'].isna()==True])
#df['gender'].isnull().sum()
```

```python
def clean_gender(x):
    if x in ['M', 'Male']:
        return 0
    elif x in ['F', 'female',"Femal"]:
        return 1
    else:
        return x
```

```python
df['gender'] = list(map(clean_gender, df['gender'])) 
```

```python
df['gender']=df['gender'].astype(int)
```
